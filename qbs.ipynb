```python
import pandas as pd

qb_adp_2022 = pd.read_csv("/Users/willpowell/Desktop/Fantasy-Football-ML/Fantasy-Football-ML_Model/ADP2022/QB.csv")
qb_adp_2022 = qb_adp_2022[['Player', 'AVG']]

qb_proj_2022 = pd.read_excel("/Users/willpowell/Desktop/Fantasy-Football-ML/Fantasy-Football-ML_Model/Projections2022/QB_proj_2022.xlsx", header=2)
qb_proj_2022 = qb_proj_2022.rename(columns={'PLAYER': 'Player'})
qb_proj_2022.columns
qb_proj_2022['Player'] = qb_proj_2022['Player'].str.split().str[:-1].str.join(' ')

qb_2022_points = pd.read_csv("/Users/willpowell/Desktop/Fantasy-Football-ML/Fantasy-Football-ML_Model/Leaders2022/QB.csv")
qb_2022_points = qb_2022_points[["Player", "TTL"]]

qbs = qb_proj_2022.merge(qb_adp_2022, on='Player', how='outer')
qbs.head()
qbs = qbs.merge(qb_2022_points, on='Player', how='outer')


#qbs_cleaned = qbs[['ATT', 'CMP', 'YDS', 'TDS', 'INTS', 'ATT.1', 'YDS.1', 'TDS.1',
 #      'FL', 'FPTS', 'TTL']]
qbs_cleaned = qbs.dropna()
qbs_cleaned
```

```python
# Hyperparameter tuning
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
import math


def fit_xgb_regressor(df_weight, lr, d, gam, a, sub, spw):
    
    X = qbs_cleaned.iloc[:, 1:-1]
    y = qbs_cleaned.iloc[:, -1]

    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
        
    
    model = XGBRegressor(max_depth=d, subsample=sub, scale_pos_weight=spw, learning_rate=lr, gamma=gam, 
                          alpha=a, random_state=2)
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)
    
    # Seeing how accurate projections were last year
    train_MSE = np.square(np.subtract(y_train, model.predict(X_train))).mean() 
    
    train_RMSE = math.sqrt(train_MSE)
    print("Train Root Mean Square Error:\n")
    print(train_RMSE)

    # Seeing how accurate projections were last year
    MSE = np.square(np.subtract(y_pred, y_test)).mean() 
    
    RMSE = math.sqrt(MSE)
    print("Root Mean Square Error:\n")
    print(RMSE)
        

    return X_train, X_test, y_train, y_test, model, RMSE, train_RMSE


def xgboost_grid_search():
    '''
    Personal function written to perform xgboost grid search
    '''
    learning_rates = [0.01, 0.03, 0.1]
    max_depths = [2, 3, 4, 5, 6]
    gammas = [0, .1]
    alphas = [0, 0.1, 0.5]
    subsamples = [0.5, 0.75, 1]

    train_scores = []
    scores = []
    params = []
    spw=0.8
    for lr in learning_rates:
        for d in max_depths:
            for gam in gammas:
                for a in alphas:
                    for sub in subsamples:
                        #print(lr,d, gam, a, sub, spw)
                        X_train, X_test, y_train, y_test, model, RMSE, train_RMSE = fit_xgb_regressor(qbs_cleaned,lr, 
                                                                                           d, gam, a, sub, spw)
                        train_scores.append(train_RMSE)
                        scores.append(RMSE)
                        params.append([lr, d, gam, a, sub, spw])

    print('max_score: ', min(scores))
    scores = np.array(scores)
    print('best parameters: ', params[np.argmin(scores)])
    print('train RMSE', train_scores[np.argmin(scores)])
    
    return params[np.argmin(scores)]

best_params = xgboost_grid_search()
```
